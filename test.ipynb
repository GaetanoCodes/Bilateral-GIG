```python
import numpy as np
from scipy.special import kv, hankel1
import matplotlib.pyplot as plt
```

```python
def phi(x, time, a1,b1,p1,a2,b2,p2):
    fourier =  kv(p1, (b1*(a1-2*1j*x))**0.5)*kv(p2, (b2*(a2+2*1j*x))**0.5)/(kv(p1, (b1*a1)**0.5)*kv(p2, (b2*a2)**0.5))
    return fourier**time
def get_dens_fourier(x_array,time, a1,b1,p1,a2,b2,p2 ):
    print("here")
    dens = []
    for x in x_array:
        u = np.arange(-50,50,0.01)
        step_u = np.mean(np.diff(u))
        value_x = (np.exp(-1j*u*x)*phi(u, time, a1,b1,p1,a2,b2,p2)).sum()*step_u
        dens.append(value_x)
    print("fourier done")
    return np.real(dens)/(2*np.pi)
```

```python
x = np.arange(-20,20,0.1)
dens = get_dens_fourier(x,1,10,2,3,1,2,-5)
plt.plot(x, dens)
print(dens.sum()*np.mean(np.diff(x)))
```
